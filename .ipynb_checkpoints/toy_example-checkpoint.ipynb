# Searching for hotspots in the two-circle dataset

This notebook implements an example analysis of hotspot detection for the mapper algorithm. 

```python
%matplotlib inline
from sklearn.datasets import make_circles
import numpy as np
import matplotlib.pyplot as plt
import sklearn.cluster as sklc
import hotmapper as hm
from itertools import chain
import pandas as pd
```

## Data 

We generate an 2-D example of two concentric circles, downloaded from [sklearn](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_circles.html). We want to be able to generate a Mapper graph that represents the core structural features of this dataset (two disconnected loops). 

```python
#import dataset
X, y = make_circles(n_samples=5000, noise=0.03, factor=0.3, random_state = 123)
```

```python
#plot the data to visualise the underlying shape 
plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c='k', marker='o', s=10, alpha=0.60)
```

We can generate an artificial hotspot in the dataset by taking a random sample, and classifying all samples that lie in the radius r of the random sample as the hotspot class. 

```python
y = hm.utils.generate_artificial_hotspot(X, 0.5, random_seed = 12)
```

```python
# plot graph with appropriate labels 
# subset dataset to y=1 and y=0 to describe legend 
bool_list = list(map(bool,y))
bool_list = list(map(bool,y))
h_class = X[bool_list]
nh_class = X[~np.array(bool_list)]

plt.figure(figsize=(6, 6))
h = plt.scatter(h_class[:,0], h_class[:,1], c='k', marker='o', s=10, alpha=0.60)
nh = plt.scatter(nh_class[:,0], nh_class[:,1], c='k', marker='o', s=10, alpha=0.60)
h = plt.scatter(h_class[:,0], h_class[:,1], c='yellow', marker='o', s=10, alpha=0.60)
plt.legend((h, nh),('Hotspot', 'Non-hotspot'),fontsize=12, loc = 1, framealpha =0.95)
plt.title("Artificial hotspot - 2D Circles")
```

## Build Mapper graph

The workflow for building the Mapper graph on a dataset is as follows:

1. Project the dataset into a low-dimensional space using the *lens* function
2. Build a cover on the lens values. The cover divides the lens function into overlapping intervals 
3. Run a clustering algorithm on the collection of data samples contained within each interval 
4. Construct a graph by representing each cluster as a node 
5. If two clusters contain overlapping samples, define an edge between the corresponding nodes 
6. Colour the graph by an *attribute* function of interest

We can manually build the Mapper algorithm using parameters identified after extensive exploration
The lens function selected is l2-norm

```python
mapper = hm.mapper.MapperGraph(data = X, 
                                lens_function = np.linalg.norm(X, axis =1), 
                                intervals = 7, 
                                overlap = 0.2,
                                clustering_algorithm = sklc.AgglomerativeClustering(n_clusters = 6, linkage = "ward"))

mapper.build_graph() 

hm.visualisation.draw_graph(mapper_graph = mapper.graph, 
              attribute_function = y, 
              samples_in_nodes = mapper.samples_in_nodes,
              size = 5, 
              style = 3, 
              tick_labels = True,
              labels = False,
              col_legend_title = "Proportion of \n Artificial Class")
```

##  Run hotspot detection on graph 

We can run the cluster detection and classification steps on the manually created Mapper graph to identify any hotspots of interest



```python
hm.visualisation.draw_graph(mapper_graph = mapper.graph, 
              attribute_function = y, 
              samples_in_nodes = mapper.samples_in_nodes,
              size = 5, 
              style = 3, 
              tick_labels = True,
              labels = True,
              col_legend_title = "Proportion of \n Artificial Class")
```

```python
#investigate each connected subgraph of the mapper graph seperately
hotspot_search = hm.hotspot.HotspotSearch(mapper_graph = mapper.graph,
                                         attribute_function = y, 
                                         samples_in_nodes = mapper.samples_in_nodes)


hotspot_nodes = hotspot_search.search_graph(attribute_threshold = 0.5, plot_dendrogram = True)

# all the results for each subgraph can be found in the hotspot variable 
print(f"Nodes identified as hotspot: {list(chain.from_iterable(hotspot_nodes))}")
```

## Searching Lens

Alternatively, the hotspot detection algorithm can be used to automatically select parameters.

This involves the selection of:
- the lens space
- set of intervals
- set of overlap
- clustering algorithm
- attribute function
- epsilon parameter (threshold at which to accept a hotspot)
- minimum sample size for a hotspot


```python
#initialise the search class with the Mapper 
search = hm.automated_parameter_search.Search(X, runs = 5)
parameters = {"non_zero_lens_features" : X.shape[1], #we want to use all the features in the lens function
              "interval_list" : [5,10,15,20,25] ,
              "overlap_list" : [0.4,0.6,0.8],
              "clustering_algorithm" : sklc.AgglomerativeClustering(n_clusters = 6, linkage = "ward"),
              "attribute_function" : y,
              "epsilon" : 0.5,
              'min_samples' : 30,
              'extreme': "higher"}
```

```python
search.build_graphs(parameters)
```

We can use the information obtained from the hotspot search to build the Mapper graph with the identified lens. We can explore the different representations of the Mapper graph from the successful interval and overlap results

```python
successful_cover = {f"Cover_{i+1}":k for i,k in enumerate(search.parameters.keys())}
print(successful_cover)
```

```python
#use the lens function identified in the search, and can observe different graphs obtained from the different successful covers
linear_lens = hm.random_lens.Lens(X, nonzero_features= parameters["non_zero_lens_features"], weights = search.parameter_lens['weights'], feature_list = search.parameter_lens["feature_list"])
i = successful_cover["Cover_12"][0]
o = successful_cover["Cover_12"][1]
```

```python
#initialise the mapper class
mapper = hm.mapper.MapperGraph(data = X, 
                        lens_function = linear_lens["lens"],
                        intervals = i, 
                        overlap =  o,
                        clustering_algorithm = parameters["clustering_algorithm"])


mapper.build_graph() 

hm.visualisation.draw_graph(mapper_graph = mapper.graph, 
              attribute_function = parameters["attribute_function"], 
              samples_in_nodes = mapper.samples_in_nodes,
              size = 5, 
              style = 2, 
              tick_labels = True,
              labels = False,
              col_legend_title = "Proportion of \n Artificial Class")
```

```python
#investigate each connected subgraph of the mapper graph seperately
hotspot_search = hm.hotspot.HotspotSearch(mapper_graph = mapper.graph,
                                         attribute_function = parameters["attribute_function"], 
                                         samples_in_nodes = mapper.samples_in_nodes)


hotspot_nodes = hotspot_search.search_graph(attribute_threshold = parameters["epsilon"], 
                                            min_sample_size = parameters["min_samples"],
                                            attribute_extreme = parameters["extreme"],
                                            plot_dendrogram = True)

# all the results for each subgraph can be found in the hotspot variable 
h_nodes = list(chain.from_iterable(hotspot_nodes))
print(f"Nodes identified as hotspot: {h_nodes}")
```

```python
hm.visualisation.draw_graph(mapper_graph = mapper.graph, 
              attribute_function = parameters["attribute_function"], 
              samples_in_nodes = mapper.samples_in_nodes,
              size = 5, 
              style = 2, 
              hotspot_nodes = h_nodes,
              tick_labels = True,
              labels = False,
              col_legend_title = "Proportion of \n Artificial Class")
```

save all the information so it can be recreated, as a new lens function is built everytime the search is run. 

- weights: the assigned weight to each feature in the lens function
- feature_list: the order of features for each weight. This is relevant when zero weights are specified for feature selection
- cover: the successful parameter combinations containing a hotspot
